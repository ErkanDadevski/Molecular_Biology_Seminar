```python
# ===================================================================
# Optimized DNA-BERT AMR Prediction Notebook (Colab-ready)
# Dataset: Staphylococcus aureus - Cefoxitin (PBP4)
# Assumptions:
#  - Sequences are raw DNA strings (A/C/G/T)
#  - Labels: 0 = non-resistant, 1 = resistant
#  - Running on Google Colab (GPU available)
# ===================================================================

# ---------------------------
# Step 0: Install dependencies (run once in Colab)
# ---------------------------
# Uncomment and run if packages aren't installed in your Colab runtime
# !pip install -q torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121
# !pip install -q transformers accelerate datasets scikit-learn biopython seaborn matplotlib tqdm

# ---------------------------
# Step 1: Imports and settings
# ---------------------------
import os
import sys
import time
import random
import math
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm

from sklearn.metrics import (accuracy_score, classification_report,
                             confusion_matrix, roc_auc_score, roc_curve)

# Device and deterministic seeds
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if DEVICE == "cuda":
    torch.cuda.manual_seed_all(SEED)

print(f"Device: {DEVICE}")

# ---------------------------
# Step 2: Download repo & load dataset (same as your original)
# ---------------------------
directory = "seminar-dlmb-2025-winter-public"
if os.path.exists(directory) and os.listdir(directory):
    print(f"✓ Directory {directory} exists and is non-empty.")
else:
    print(f"Downloading {directory}...")
    os.system('git clone https://github.com/hzi-bifo/seminar-dlmb-2025-winter-public.git')

sys.path.append('seminar-dlmb-2025-winter-public/src/')
try:
    from amr.amr_utility import get_seq_label_simple
except Exception as e:
    raise RuntimeError("Could not import amr_utility. Make sure the repo is present.") from e

print("\n" + "=" * 60)
print("LOADING STAPHYLOCOCCUS AUREUS - CEFOXITIN DATASET")
print("=" * 60)
ds = get_seq_label_simple("Staphylococcus_aureus_cefoxitin_pbp4")

seq_train = [x[0] for x in ds["train"]]
y_train = [int(x[1]) for x in ds["train"]]

seq_test = [x[0] for x in ds["test"]]
y_test = [int(x[1]) for x in ds["test"]]

print(f"✓ Training samples: {len(seq_train)}")
print(f"✓ Test samples: {len(seq_test)}")
print(f"✓ Example sequence length: {len(seq_train[0])} bp")
print(f"✓ Resistant samples (train): {sum(y_train)}/{len(y_train)}")
print(f"✓ Resistant samples (test): {sum(y_test)}/{len(y_test)}")
print("=" * 60)

# ---------------------------
# Step 3: Quick exploration & plot (optional)
# ---------------------------
def explore_dataset(seq_train, y_train, seq_test, y_test, save_plot=True):
    len_train = [len(s) for s in seq_train]
    len_test = [len(s) for s in seq_test]

    print("\nSequence Length Statistics:")
    print(f"  Train - Min: {min(len_train)}, Max: {max(len_train)}, Median: {np.median(len_train):.0f}")
    print(f"  Test  - Min: {min(len_test)}, Max: {max(len_test)}, Median: {np.median(len_test):.0f}")

    y_train_int = list(map(int, y_train))
    y_test_int = list(map(int, y_test))

    print("\nLabel Distribution:")
    print(f"  Train - Resistant: {sum(y_train_int)} ({100*sum(y_train_int)/len(y_train_int):.1f}%)")
    print(f"  Test  - Resistant: {sum(y_test_int)} ({100*sum(y_test_int)/len(y_test_int):.1f}%)")

    # nucleotide composition (sample)
    sample_seq = ''.join(seq_train[:100])
    counts = {nt: sample_seq.count(nt) for nt in 'ACGT'}
    total = sum(counts.values())
    print("\nNucleotide composition (sample of first 100 seqs):")
    for k, v in counts.items():
        print(f"  {k}: {100*v/total:.1f}%")

    # plots
    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    axes[0].hist([len_train, len_test], label=['Train', 'Test'], bins=20, alpha=0.7)
    axes[0].set(title='Sequence length', xlabel='bp')

    labels = ['Non-resistant', 'Resistant']
    train_counts = [len(y_train_int)-sum(y_train_int), sum(y_train_int)]
    test_counts = [len(y_test_int)-sum(y_test_int), sum(y_test_int)]
    x = np.arange(len(labels))
    width = 0.35
    axes[1].bar(x - width/2, train_counts, width, label='Train')
    axes[1].bar(x + width/2, test_counts, width, label='Test')
    axes[1].set_xticks(x); axes[1].set_xticklabels(labels); axes[1].set(title='Label distribution')

    axes[2].bar(list(counts.keys()), [100*v/total for v in counts.values()])
    axes[2].set(title='Nucleotide % (sample)')
    axes[2].set_ylim(0, 100)

    plt.tight_layout()
    if save_plot:
        plt.savefig('dataset_exploration.png', dpi=200, bbox_inches='tight')
        print("✓ Saved dataset_exploration.png")
    plt.show()

explore_dataset(seq_train, y_train, seq_test, y_test)

# ---------------------------
# Step 4: Faster Dataset + Tokenization
# ---------------------------
# Config - tuned for Colab GPU
CONFIG = {
    "MODEL_NAME": "zhihan1996/DNABERT-2-117M",
    "MAX_LENGTH": 256,        # good tradeoff on Colab
    "BATCH_SIZE": 4,          # per step
    "ACCUM_STEPS": 4,         # effective batch = 16 (BATCH_SIZE * ACCUM_STEPS)
    "NUM_EPOCHS": 3,
    "LR": 1e-5,
    "UNFREEZE_LAST_N": 4,     # last N backbone layers to unfreeze
    "FP16": True,             # use mixed precision on GPU
    "WEIGHTED_LOSS": True,
    "EARLY_STOP_PATIENCE": 2,
    "SAVE_PATH": "best_dnabert_amr_colab.pth",
    "KMER_AUGMENT": False     # set True only if backbone tokenizer requires kmers
}

# Try to load tokenizer (best-effort). Use use_fast=False to be safe for custom tokenizers.
print("\nLoading tokenizer (may take a minute)...")
try:
    tokenizer = AutoTokenizer.from_pretrained(CONFIG["MODEL_NAME"], use_fast=False)
    print("✓ Tokenizer loaded.")
except Exception as e:
    print(f"Tokenizer load failed: {e}\nProceeding without tokenizer (fallback tokenization will be used).")
    tokenizer = None

class DNABertDatasetFast(Dataset):
    """Dataset to feed DNA-BERT or fallback encoder"""
    def __init__(self, sequences, labels, tokenizer=None, max_length=256, k=6, kmer_augment=False):
        self.seqs = sequences
        self.labels = torch.tensor([int(x) for x in labels], dtype=torch.long)
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.k = k
        self.kmer_augment = kmer_augment

    def __len__(self):
        return len(self.seqs)

    def _to_kmer(self, seq):
        seq = seq.upper()
        kmers = [seq[i:i+self.k] for i in range(len(seq)-self.k+1)]
        return " ".join(kmers)

    def __getitem__(self, idx):
        seq = self.seqs[idx].upper()
        label = self.labels[idx]

        if self.tokenizer is not None:
            text = self._to_kmer(seq) if self.kmer_augment else seq
            enc = self.tokenizer(text, padding='max_length', truncation=True,
                                 max_length=self.max_length, return_tensors='pt')
            return {
                "input_ids": enc["input_ids"].squeeze(0),
                "attention_mask": enc["attention_mask"].squeeze(0),
                "labels": label
            }
        else:
            # fallback char encoding
            char_to_idx = {'A':1,'C':2,'G':3,'T':4,'N':0}
            ids = [char_to_idx.get(ch, 0) for ch in seq[:self.max_length]]
            if len(ids) < self.max_length:
                ids += [0] * (self.max_length - len(ids))
            return {
                "input_ids": torch.tensor(ids, dtype=torch.long),
                "attention_mask": torch.ones(self.max_length, dtype=torch.long),
                "labels": label
            }

# ---------------------------
# Step 5: Model class (partial unfreeze + stronger head)
# ---------------------------
class DNABertAMRModel(nn.Module):
    def __init__(self, model_name, num_classes=2, freeze_backbone=True, unfreeze_last_n=4, classifier_dropout=0.2):
        super().__init__()
        self.model_name = model_name

        # Try loading backbone
        try:
            print(f"Loading backbone model: {model_name} ...")
            self.backbone = AutoModel.from_pretrained(model_name)
            self.hidden_size = self.backbone.config.hidden_size
            print(f"✓ Backbone loaded. Hidden size: {self.hidden_size}")
        except Exception as e:
            print(f"Backbone load failed: {e}\nFalling back to small transformer encoder.")
            self.backbone = None
            self.hidden_size = 512
            self.embedding = nn.Embedding(6, self.hidden_size)
            self.encoder = nn.TransformerEncoder(
                nn.TransformerEncoderLayer(self.hidden_size, 8, self.hidden_size*4, batch_first=True),
                num_layers=4
            )

        # Freeze backbone first (safe)
        if freeze_backbone and self.backbone is not None:
            for p in self.backbone.parameters():
                p.requires_grad = False

        # Partial unfreeze: attempt to find encoder layers and unfreeze last N
        if unfreeze_last_n > 0 and self.backbone is not None:
            found = False
            # candidate attribute paths where encoder layers might live
            candidates = [
                ["encoder", "layer"],
                ["base_model", "encoder", "layer"],
                ["bert", "encoder", "layer"],
                ["roberta", "encoder", "layer"]
            ]
            for cand in candidates:
                try:
                    obj = self.backbone
                    for attr in cand:
                        obj = getattr(obj, attr)
                    N = len(obj)
                    for layer in obj[-unfreeze_last_n:]:
                        for p in layer.parameters():
                            p.requires_grad = True
                    print(f"Unfroze last {unfreeze_last_n} layers from {'/'.join(cand)} (total={N})")
                    found = True
                    break
                except Exception:
                    continue
            if not found:
                print("Couldn't locate encoder layers to partially unfreeze; skipping partial unfreeze.")

        # Classifier head (deeper than single linear)
        self.classifier = nn.Sequential(
            nn.Linear(self.hidden_size, self.hidden_size // 2),
            nn.ReLU(),
            nn.Dropout(classifier_dropout),
            nn.Linear(self.hidden_size // 2, max(64, self.hidden_size // 4)),
            nn.ReLU(),
            nn.Dropout(classifier_dropout / 2),
            nn.Linear(max(64, self.hidden_size // 4), num_classes)
        )

    def forward(self, input_ids, attention_mask=None):
        if self.backbone is not None:
            outputs = self.backbone(input_ids=input_ids, attention_mask=attention_mask, return_dict=True)
            # try use last_hidden_state CLS token
            if hasattr(outputs, "last_hidden_state"):
                cls_vec = outputs.last_hidden_state[:, 0, :]  # [CLS]
            else:
                # fallback to mean pooling
                cls_vec = outputs.hidden_states[-1].mean(dim=1)
        else:
            x = self.embedding(input_ids)
            cls_vec = self.encoder(x).mean(dim=1)

        logits = self.classifier(cls_vec)
        return logits

# ---------------------------
# Helpers: class weights, plotting
# ---------------------------
def compute_class_weights(labels):
    cnt = Counter(int(x) for x in labels)
    total = sum(cnt.values())
    # inverse freq weights (normalized)
    weights = []
    for cls in sorted(cnt.keys()):
        weights.append(total / (cnt[cls] + 1e-8))
    weights = np.array(weights, dtype=np.float32)
    weights = weights / weights.sum() * len(weights)
    return torch.tensor(weights, dtype=torch.float32)

def plot_history(history, save=True):
    epochs = range(1, len(history['train_loss']) + 1)
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(epochs, history['train_loss'], '-o', label='train_loss')
    plt.plot(epochs, history['val_loss'], '-o', label='val_loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.title('Loss')

    plt.subplot(1,2,2)
    plt.plot(epochs, history['train_acc'], '-o', label='train_acc')
    plt.plot(epochs, history['val_acc'], '-o', label='val_acc')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy (%)'); plt.legend(); plt.title('Accuracy')

    plt.tight_layout()
    if save:
        plt.savefig('training_history.png', dpi=200, bbox_inches='tight')
        print("✓ Saved training_history.png")
    plt.show()

# ---------------------------
# Step 6: Training loop (FP16 + gradient accumulation + early stopping)
# ---------------------------
def validate(model, val_loader, criterion, device):
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch.get('attention_mask', None)
            if attention_mask is not None:
                attention_mask = attention_mask.to(device)
            labels = batch['labels'].to(device)

            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = criterion(logits, labels)

            val_loss += loss.item()
            preds = torch.argmax(logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    return val_loss / max(1, len(val_loader)), 100 * correct / total

def train_model_fast(model, train_loader, val_loader, config, class_weights=None):
    device = CONFIG["DEVICE"] if "DEVICE" in CONFIG else DEVICE
    model = model.to(device)

    if config["WEIGHTED_LOSS"] and class_weights is not None:
        criterion = nn.CrossEntropyLoss(weight=class_weights.to(device))
    else:
        criterion = nn.CrossEntropyLoss()

    optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=config["LR"], weight_decay=0.01)

    total_steps = math.ceil(len(train_loader) * config["NUM_EPOCHS"] / config["ACCUM_STEPS"])
    try:
        scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=config["LR"], total_steps=total_steps, pct_start=0.1)
    except Exception:
        scheduler = None

    scaler = torch.cuda.amp.GradScaler(enabled=(config["FP16"] and device.startswith("cuda")))

    best_val_acc = 0.0
    no_improve = 0
    history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

    for epoch in range(config["NUM_EPOCHS"]):
        model.train()
        running_loss = 0.0
        running_correct = 0
        running_total = 0
        optimizer.zero_grad()

        pbar = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"Epoch {epoch+1}/{config['NUM_EPOCHS']}")
        for step, batch in pbar:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch.get('attention_mask', None)
            if attention_mask is not None:
                attention_mask = attention_mask.to(device)
            labels = batch['labels'].to(device)

            with torch.cuda.amp.autocast(enabled=(config["FP16"] and device.startswith("cuda"))):
                logits = model(input_ids=input_ids, attention_mask=attention_mask)
                loss = criterion(logits, labels) / config["ACCUM_STEPS"]

            scaler.scale(loss).backward()

            preds = torch.argmax(logits, dim=1)
            running_correct += (preds == labels).sum().item()
            running_total += labels.size(0)
            running_loss += loss.item() * config["ACCUM_STEPS"]

            if (step + 1) % config["ACCUM_STEPS"] == 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
                if scheduler is not None:
                    try:
                        scheduler.step()
                    except Exception:
                        pass

            pbar.set_postfix({'loss': f"{running_loss/(step+1):.4f}", 'acc': f"{100*running_correct/running_total:.2f}%"})

        train_loss = running_loss / len(train_loader)
        train_acc = 100 * running_correct / running_total

        val_loss, val_acc = validate(model, val_loader, criterion, device)

        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        print(f"\nEpoch {epoch+1} summary: Train Loss={train_loss:.4f}, Train Acc={train_acc:.2f}% | Val Loss={val_loss:.4f}, Val Acc={val_acc:.2f}%")

        # Save best model
        if val_acc > best_val_acc + 1e-4:
            best_val_acc = val_acc
            torch.save(model.state_dict(), config["SAVE_PATH"])
            print(f"✓ New best model saved (Val Acc={val_acc:.2f}%)")
            no_improve = 0
        else:
            no_improve += 1
            print(f"No improvement for {no_improve} epoch(s).")

        if no_improve >= config["EARLY_STOP_PATIENCE"]:
            print("Early stopping triggered.")
            break

    return model, history

# ---------------------------
# Step 7: Full pipeline runner
# ---------------------------
def run_pipeline(seq_train, y_train, seq_test, y_test, config=CONFIG, tokenizer=tokenizer):
    print("\n" + "="*60)
    print("RUNNING DNA-BERT AMR PIPELINE")
    print("="*60)
    device = DEVICE
    print(f"Device: {device}")

    # Dataloaders
    train_ds = DNABertDatasetFast(seq_train, y_train, tokenizer=tokenizer, max_length=config["MAX_LENGTH"], k=6, kmer_augment=config["KMER_AUGMENT"])
    val_ds = DNABertDatasetFast(seq_test, y_test, tokenizer=tokenizer, max_length=config["MAX_LENGTH"], k=6, kmer_augment=config["KMER_AUGMENT"])

    train_loader = DataLoader(train_ds, batch_size=config["BATCH_SIZE"], shuffle=True, num_workers=2, pin_memory=True if device=='cuda' else False)
    val_loader = DataLoader(val_ds, batch_size=config["BATCH_SIZE"], shuffle=False, num_workers=2, pin_memory=True if device=='cuda' else False)

    print(f"✓ Train batches: {len(train_loader)}, ✓ Val batches: {len(val_loader)}")
    print(f"Batch size (per step): {config['BATCH_SIZE']}, Accumulation steps: {config['ACCUM_STEPS']}")

    # Class weights
    class_weights = None
    if config["WEIGHTED_LOSS"]:
        class_weights = compute_class_weights(y_train)
        print("Class weights:", class_weights.numpy())

    # Model
    model = DNABertAMRModel(config["MODEL_NAME"], num_classes=2, freeze_backbone=True, unfreeze_last_n=config["UNFREEZE_LAST_N"])
    # Train
    start = time.time()
    model, history = train_model_fast(model, train_loader, val_loader, config, class_weights)
    elapsed = time.time() - start
    print(f"Total training time: {elapsed/60:.2f} minutes")

    # Load best model if saved
    if os.path.exists(config["SAVE_PATH"]):
        print("Loading best model weights...")
        model.load_state_dict(torch.load(config["SAVE_PATH"], map_location=device))

    # Final evaluation
    model = model.to(device)
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for batch in tqdm(val_loader, desc="Final evaluation"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch.get('attention_mask', None)
            if attention_mask is not None:
                attention_mask = attention_mask.to(device)
            labels = batch['labels'].to(device)

            logits = model(input_ids=input_ids, attention_mask=attention_mask)
            probs = torch.softmax(logits, dim=1)[:,1]
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    # Metrics
    acc = accuracy_score(all_labels, all_preds)
    print(f"\nFinal Test Accuracy: {acc*100:.2f}%")
    print("\nClassification report:")
    print(classification_report(all_labels, all_preds, target_names=["Non-resistant","Resistant"]))

    # Confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    print("Confusion matrix:\n", cm)

    # ROC AUC if both classes exist
    if len(set(all_labels)) == 2:
        auc = roc_auc_score(all_labels, all_probs)
        print(f"ROC AUC: {auc:.4f}")
        fpr, tpr, _ = roc_curve(all_labels, all_probs)
        plt.figure(figsize=(5,4))
        plt.plot(fpr, tpr, label=f"AUC={auc:.3f}")
        plt.plot([0,1],[0,1],'k--')
        plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC curve")
        plt.legend()
        plt.show()

    plot_history(history)
    return model, history, {"accuracy": acc, "confusion_matrix": cm}

# ---------------------------
# Step 8: Run everything
# ---------------------------
if __name__ == "__main__":
    # Final sanity check & run
    model, history, results = run_pipeline(seq_train, y_train, seq_test, y_test, config=CONFIG, tokenizer=tokenizer)
    print("\nDone. Model saved to:", CONFIG["SAVE_PATH"])
```
